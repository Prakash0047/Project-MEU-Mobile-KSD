Spilitting data into dependent and idependent form

```
X=project.iloc[:,1:].values
y=project.iloc[:,0:1].values
X.shape
y.shape
```

Data Spilliting

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
print(X_train.shape)
print(X_test.shape)
```

Importing Classifier Library

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import VotingClassifier
clf1=RandomForestClassifier(n_estimators=100)
clf2=BaggingClassifier(base_estimator=KNeighborsClassifier(), n_estimators=100)
clf3=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=100)
clf4=VotingClassifier(estimators=[('Rf',clf1),('BAG',clf2),('ADA',clf3)],
                      weights=[1,1,1],
                      voting='hard')
```

Accuracy Score

```
from sklearn.metrics import accuracy_score
clf=[clf1,clf2,clf3,clf4]
clf_name=['RF','BAG','ADA','ENSEMBLE']
for cl,name in zip(clf,clf_name):
  cl.fit(X_train,y_train)
  y_train_pred=cl.predict(X_train)
  y_test_pred=cl.predict(X_test)
  train_acc=accuracy_score(y_train,y_train_pred)
  test_acc=accuracy_score(y_test,y_test_pred)
  print('%s:- %.2f %.2f'%(name,train_acc,test_acc))
```

```
from sklearn.model_selection import cross_val_score
clf=[clf1,clf2,clf3,clf4]
clf_name=['RF','BAG','ADA','ENSEMBLE']
acc=[]
for cl,name in zip(clf,clf_name):
  scores=cross_val_score(estimator=cl,cv=5,X=X,y=y,scoring='accuracy')
  print('Accuracy: %.2f +/- %.2f [%s]'%(scores.mean(),scores.std(),name))
  acc.append(scores.mean())
```
